# import

```python
import sys
import os
import pandas as pd
import seaborn as sns
import numpy as np
from src.constants import MODEL_CONST
from src.init_DFs.per_game import PerGameInit
initPG = PerGameInit()
```

# configure

```python
season = "14_16"
situation = "5on5"
```

```python
# Configure PER_GAME_{SITUATION}_{SEASON}.csv
games = pd.read_csv(MODEL_CONST['CSV_DB_PATH'] + f"{season}/games_{season}.csv")
team_5on5 = pd.read_csv(MODEL_CONST['CSV_DB_PATH'] + f"{season}/team_x_game_{situation}_{season}.csv")
odds = pd.read_csv(MODEL_CONST['CSV_DB_PATH'] + f"{season}/odds_x_game_{season}.csv")

# merge team_{SIT}, games, odds
pg = pd.merge(team_5on5, games)
pg = pd.merge(pg, odds)

# create PER_GAME_{SITUATION}_{SEASON}.csv
pg = initPG.init_per_game(pg)
pg.to_csv(MODEL_CONST['CSV_TEMP_PATH'] + f'PER_GAME_{situation}_{season}.csv', header=True, index=False)

# create PER_GAME_BY_TEAM_{SITUATION}_{SEASON}.csv
pg_copy = pg.copy()
pg_teams = initPG.init_by_team(pg, pg_copy)
pg_teams.to_csv(MODEL_CONST['CSV_TEMP_PATH'] + f'PER_GAME_BY_TEAM_{situation}_{season}.csv', header=True, index=False)
pg_teams
```
