```python
# Import Dependencies
import sys
import pandas as pd
import seaborn as sns
import numpy as np
from src.init_DFs.per_game import PerGameInit
perGameInit = PerGameInit()
```

```python
# Configure PER_GAME_{SITUATION}_{SEASON}.csv
season = "2022"
situation = "5on5"
csv_database = "C:/wamp64/www/bet-nhl-model/csv_database/"
games = pd.read_csv(csv_database + f"{season}/games_{season}.csv")
team_5on5 = pd.read_csv(csv_database + f"{season}/team_x_game_{situation}_{season}.csv")
odds = pd.read_csv(csv_database + f"{season}/odds_x_game_{season}.csv")

# merge team_{SIT}, games, odds
pg = pd.merge(team_5on5, games)
pg = pd.merge(pg, odds)

# create PER_GAME_{SITUATION}_{SEASON}.csv
pg = perGameInit.init_per_game(pg)
pg.to_csv(f'PER_GAME_{situation}_{season}.csv', header=True, index=False)

# create PER_GAME_BY_TEAM_{SITUATION}_{SEASON}.csv
pg_copy = pg.copy()
pg_teams = perGameInit.init_by_team(pg, pg_copy)
pg_teams.to_csv(f'PER_GAME_BY_TEAM_{situation}_{season}.csv', header=True, index=False)
pg_teams
```
